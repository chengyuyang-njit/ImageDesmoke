```python
import os
import shutil
from PIL import Image
import numpy as np
from scipy.io import savemat
from skimage.metrics import structural_similarity as ssim, peak_signal_noise_ratio as psnr
```

```python
source_folder = "/mmfs1/project/cliu/cy322/projects/pytorch-CycleGAN-and-pix2pix/results/desmoke-pixel2pixel/test_latest/images"
destination_folder = "/mmfs1/project/cliu/cy322/projects/ImageDesmoke/saved/evaluation/cycle_gan/outputs"
```

```python
for filename in os.listdir(source_folder):
    if 'fake' in filename:
        img = Image.open(os.path.join(source_folder, filename))
        resized_image = img.resize((700,350))
        resized_image.save(os.path.join(destination_folder,filename))
```
